```python
import cx_Oracle
import pandas as pd
import requests
from xml.etree.ElementTree import fromstring
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc
import pyproj
import folium
from folium.features import CustomIcon
import time
from IPython import display
```

```python
conn = cx_Oracle.connect('open_source/1111@localhost:1521/xe')
cur = conn.cursor()
```

```python
def load_data() :
    cur.execute("""
    select column_name from user_tab_columns where table_name = 'BUS_753'
    """)
    column_list = [record[0] for record in cur]
    column_list

    cur.execute('select * from bus_753 where 제공시각 = (select max(제공시각) from bus_753)')
    bus753_df= pd.DataFrame(cur, columns=column_list)
    
    font_location = "C:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=font_location).get_name()
    rc('font',family=font_name)
    plt.rcParams['axes.unicode_minus'] = False
    bus_station_df = pd.read_csv('서울시버스정류소좌표데이터.csv', encoding='cp949')
    
    bus_station_df=bus_station_df[['정류소ID','X좌표','Y좌표']]
    bus753_df['정류소ID']=bus753_df['정류소ID'].astype('int')
    pd.merge(bus753_df , bus_station_df, how = "inner", on = '정류소ID')
    merge_df=pd.merge(bus753_df, bus_station_df,on='정류소ID')
    drop_dup_df=merge_df.drop_duplicates(['정류소ID'])
    coord = np.array(drop_dup_df[['Y좌표', 'X좌표']])
    
    
    return drop_dup_df
```

```python
load_data()
```

```python
def create_map(drop_dup_df) :
    coord = np.array(drop_dup_df[['Y좌표', 'X좌표']])
    
    
    center = [37.57167752, 126.9096099]
    m = folium.Map(location=center, zoom_start=15)
    folium.PolyLine(locations = coord,color='black').add_to(m)
    
    return m
```

```python
m=create_map(load_data())
m
```

```python
def plt_update():

    columns = ['정류소ID','정류소명', '첫번째도착정보메시지', '두번째도착정보메시지',
               '제공시각','X좌표','Y좌표']
    df = pd.DataFrame(load_data(),columns=columns)
    
    
    for i in range(len(load_data())):
        icon_image1 = 'BUS.png'
        icon1 = folium.features.CustomIcon(icon_image1, icon_size=(50, 50))
        icon_image2 = 'bus_stop.png'
        icon2 = folium.features.CustomIcon(icon_image2, icon_size=(30, 45))
        if(load_data().iloc[i]['첫번째도착정보메시지']!='곧 도착'):
            folium.Marker([load_data().iloc[i][6], load_data().iloc[i][5]], popup=load_data().iloc[i][1],icon=icon2).add_to(m)
        else :
            folium.Marker([load_data().iloc[i][6], load_data().iloc[i][5]], popup=load_data().iloc[i][1],icon=icon1).add_to(m)
            
    return m
```

```python
hdisplay = display.display("", display_id=True)
is_first = True
for f in range(1):

    load_data()
    plt_update()
    
    if is_first == True:
        is_first = False
    else :
        time.sleep(15)
        
    hdisplay.update(m)
```
