```python
import numpy as np
import pandas as pd
import cx_Oracle

import folium
from folium.plugins import MarkerCluster
from folium import Marker
import time

from IPython import display
```

```python
con = cx_Oracle.connect('open_source/1111@localhost:1521/xe')
cur = con.cursor()
```

```python
def load_data() :
    cur.execute('select * from bike_real_time where 제공시각 = (select max(제공시각) from bike_real_time)')

    columns = ['대여소이름','거치대개수','남은자전거수','거치율','위도','경도','대여소ID','제공시각']
    bike_real_time = pd.DataFrame(cur,columns = columns)
    return bike_real_time

#bike_real_time
```

```python
def create_map(bike_real_time) :
    # 합정동기준
    latitude = 37.552746
    longitude = 126.918617
    
    bike_map = folium.Map(location = [latitude, longitude],
               zoom_start=15, 
               width=750, 
               height=500
              )
    return bike_map
```

```python
m=create_map(load_data())
#m
```

```python
def plt_update():
    for name, lat, lng, m_num  in zip(bike_real_time['대여소이름'],bike_real_time['위도'],bike_real_time['경도'],bike_real_time['남은자전거수']):
        Marker(location =[lat,lng],
               popup=m_num,
               tooltip=name,
               icon=folium.Icon(color='blue',icon='bookmark')
               ).add_to(bike_map)
```

```python
hdisplay = display.display("", display_id=True)
is_first = True
for f in range(3):

    load_data()
    plt_update()
    
    if is_first == True:
        is_first = False
    else :
        time.sleep(20)
        
    hdisplay.update(bike_map)
```
