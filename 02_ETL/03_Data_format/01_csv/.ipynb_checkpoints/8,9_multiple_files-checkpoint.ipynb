```python
import pandas as pd
import csv
import glob
import os

input_path = '.'
ouput_file = 'output_files/9output_basic_pandas.csv'

files_list = []
all_files = glob.glob(os.path.join(input_path,'sales_*'))

for file in all_files:
    data = pd.read_csv(file)
    files_list.append(data)
    

result_file = pd.concat(files_list, axis = 0)
result_file
```

```python
result_file.to_csv(ouput_file, index = False)
```
