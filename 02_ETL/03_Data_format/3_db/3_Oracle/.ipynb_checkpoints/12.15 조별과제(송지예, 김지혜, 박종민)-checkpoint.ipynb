# 2. 첫번째 분석된 데이터를 기반으로 각 시설별 장애인을 위한 편의시설을 검색

```python
convenient_list = ['장애인', '휠체어','노약자','노인','임산부']
test = '|'.join(convenient_list)
```

```python
cf = df[df['편의시설정보'].str.contains(test)]
```

```python
cf
```

```python
len(cf)
```

##### (국 공 사 대)립 별 장애인을 위한 편의시설 보유 현황 

```python
convience_total = []
cf.groupby('박물관미술관구분')['박물관미술관구분'].count()
```

```python
cf_1 = list(cf.groupby('박물관미술관구분')['박물관미술관구분'].count())
cf_1
```

```python
for i in cf_1 :
    convience_total.append(i)
convience_total
```

```python
convience_total.append(df['박물관미술관구분'].count().sum()-54)
convience_total
```

```python
convience_total_df = pd.DataFrame(convience_total)
convience_total_df
```

```python
convience_total_df.index = ['공립', '국립', '대학','사립','전체']
```

```python
convience_total_df
```

```python
plt.title('구분별 장애인을 위한 편의시설 확보여부',position=(0, 0), 
          fontsize=13)
explode = [2, 1, 0, 2, 0]
labels = ['공립', '국립', '대학','사립','편의시설미보유']
colors = ['skyblue', 'red', 'orange', 'limegreen','lightgray']

plt.pie(convience_total_df[0],labels=labels, explode=explode,
         autopct='%.1f%%', colors = colors,
       startangle=180)
plt.show()
```

## 4.성인입장료 현황을 알아보기 위해 가격을 그룹으로 묶어서 검색

```python
data = museum.groupby('어른관람료')['어른관람료'].count()
data
```

#### => 성인입장료 현황을 알아보기 위해 가격을 그룹으로 묶어서 검색하였다.

```python
ad_0 = museum.loc[museum['어른관람료'] == 0]
data_0 = ad_0.groupby('어른관람료')['어른관람료'].count().sum()
ad_1000 = museum.loc[(museum['어른관람료']<=1000)&(museum['어른관람료']>0)]
data = ad_1000.groupby('어른관람료')['어른관람료'].count().sum()
ad_5000 = museum.loc[(museum['어른관람료']<=5000)&(museum['어른관람료']>1000)]
data2 = ad_5000.groupby('어른관람료')['어른관람료'].count().sum()
ad_10000 = museum.loc[(museum['어른관람료']<=10000)&(museum['어른관람료']>5000)]
data3 = ad_10000.groupby('어른관람료')['어른관람료'].count().sum()
ad_20000 = museum.loc[(museum['어른관람료']<=20000)&(museum['어른관람료']>10000)]
data4 = ad_20000.groupby('어른관람료')['어른관람료'].count().sum()
ad_max = museum.loc[museum['어른관람료']>20000]
data5 = ad_max.groupby('어른관람료')['어른관람료'].count().sum()
```

#### => 금액별 분류를 해주었다

```python
import numpy as np
x=np.arange(6)
colors = ['#ffc000', '#ff9999', '#8fd9b6', '#d395d0', 'pink', 'cyan']
explode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05]
price = ['무료','1천이하','5천이하','1만이하','2만이하','2만이상']
values = [data_0,data,data2,data3,data4,data5]
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)  
plt.bar(x,values,color=colors)
plt.xticks(x,price)
plt.xlabel('가격')
plt.ylabel('박물관 개수')
plt.title('전국 박물관 성인요금 현황')

plt.subplot(1, 2, 2)  
plt.pie(values,labels=price,autopct='%.1f%%', colors=colors, explode=explode)
plt.title('전국 박물관 성인요금 비율')

plt.show()
```

#### => 금액별 분류를 시각화 하였다
