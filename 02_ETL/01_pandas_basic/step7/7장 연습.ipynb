```python
grade = pd.DataFrame({'name':['jack','eva','john','happy'],
                      'math':[90,85,88,76], 
                      'english':[75,83,92,81], 
                      'science':[60,90,77,83], 
                      'class':['a','a','b','c']})
grade
```

```python
def my_grade(math,english,science):
    return (math + english + science) / 3
```

```python
grade['avg'] = grade.apply(lambda grade: my_grade(grade['math'],grade['english'],grade['science']), axis = 1)
```

```python
grade = grade.round(1)
```

```python
def mytest_pass(avg):
    return grade['avg'] > 80
```
