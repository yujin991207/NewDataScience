```python
# 결측치를 '-' 로 저장한 데이터가 있는지 모든 열을 대상으로 검색
# 첫번째 행부터 '-' 있는지 검색한다.
for name in HeroInfo.columns:
    print(HeroInfo[HeroInfo[name]=='-'])
```

```python
# Aliases 열의 모든 행의 값에 대해 '-'문자와 비교한 결과를 시리즈로 반환
(HeroInfo['Aliases']=='-')
```

```python
HeroInfo.columns.size
```

```python
# 전체 column 중 '-' 있는 데이터를 본다.
# 그리고 '-' 있는 데이터를 na_column_list에 append한다.
# if True in .. '-' 포함 조건이 만족한다면 .. 
# 해당 열 이름에 '-'이 있다는 말이다.
na_column_list=[]
for name in HeroInfo.columns:
    if True in (HeroInfo[name]=='-').tolist():
        if name not in na_column_list:
            na_column_list.append(name)
len(na_column_list)
```

```python
na_column_list
```

```python
# '-'을 NaN으로 바꿈.
for i in range(0,748) :
    a = HeroInfo.iloc[i].str.replace('-','NaN')
    HeroInfo.iloc[i] = a
```

```python
HeroInfo = HeroInfo.fillna('NaN')
```

```python
HeroInfo.head(1)
```

```python
HeroInfo.columns
```

```python
# 데이터프레임에서 사용하고 있는 메모리현황을 보여준다.
HeroInfo.memory_usage(deep=True).head()
```

```python
na_column_list
```

```python
HeroInfo.head(1)
```

```python
# 위에서 식별한 결측치가 있는 열 중에서 수치형 데이터를 포함한 데이터를 식별한다. 
cols = ['Character Score', 'Intelligence', 'Strength', 'Speed','Durability', 'Power', 'Combat']
```

```python
# ['Character Score', 'Intelligence', 'Strength', 'Speed','Durability', 'Power', 'Combat'] 열에 한하여 
# 결측치를 0으로 처리
# 위 해당 열을 float형으로 변환
for c in cols :
    HeroInfo[c] = HeroInfo[c].astype('float')
    HeroInfo[c] = HeroInfo[c].fillna(0)
```

```python
HeroInfo.dtypes
```

```python
HeroInfo.isnull().sum().head(24)
```

```python
aa = HeroInfo.filter(items=[ 'Character Score', 'Intelligence',
                           'Strength','Speed','Durability','Power','Combat'])
```

```python
aa.shape
```

```python
aa.head()
```

```python
aa = aa.sort_values(['Character Score'],ascending=False)
```

```python
aa.nlargest(100, 'Character Score').head()
```

```python
# aa.nlargest(100, 'Character Score')의 결과를 대상으로 nsmallest(10,'Combat')
aa.nlargest(100, 'Character Score').nsmallest(10,'Combat')
```
