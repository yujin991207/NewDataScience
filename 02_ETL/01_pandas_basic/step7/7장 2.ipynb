## 예제 7-6 추가 메소드 적용하기

### 누적값/누적곱
* cumsum
* cumprod

```python
import numpy as np
```

```python
df = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],'관측값_B':[3,4,np.NaN,9,2,5]})
```

```python
df
```

```python
# 원본을 기준으로 각 행마다 누적 합을 반환한다.
# NaN이 있는 행은 해당 행에서만 누적값을 NaN으로 처리하고 그 이외에는 누적대상에서 제외한다.
df.cumsum()
```

```python
df['관측값_A_누적값']=df['관측값_A'].cumsum()
df
```

```python
df['관측값_B_누적합']=df['관측값_B'].cumsum()
df
```

```python
df = df[['관측값_A','관측값_A_누적값','관측값_B','관측값_B_누적합']]
df
```

```python
df = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],'관측값_B':[3,4,np.NaN,9,2,5]})
df
```

```python
# get_loc(열이름)을 지정하면 해당 열의 수치형 열 인덱스를 반환 
# 관측값_A열 다음에 insert하기 위해 1을 더함.
df.insert(df.columns.get_loc('관측값_A')+1,'관측값_A_누적값',df['관측값_A'].cumsum())
df
```

```python
df.insert(df.columns.get_loc('관측값_B')+1,'관측값_B_누적값',df['관측값_B'].cumsum())
```

```python
df
```

```python
# 원본을 기준으로 누적곱을 반환한다
# 누적곱: 특정 알고리즘에서 사용한다. 
# 파이썬에서 오버플로우 테스트를 하기 위해 다음 예제에서 사용
df['관측값_A'].cumprod(), df['관측값_B'].cumprod()
```

```python
df.insert(df.columns.get_loc('관측값_A_누적값')+1,'관측값_A_누적곱',df['관측값_A'].cumprod())
df
```

```python
df.insert(df.columns.get_loc('관측값_B_누적값')+1,'관측값_B_누적곱',df['관측값_B'].cumprod())
df
```

```python
if (1 + 2) == 3:
    print("1 + 2는 당연히 3입니다!")
else:
    print("1 + 2는 3이 아닙니다!!")
```

```python
if (0.1 + 0.2) == 0.3:
    print("0.1 + 0.2는 당연히 0.3입니다.")
else:
    print("0.1 + 0.2는 0.3이 아닙니다!!")
```
