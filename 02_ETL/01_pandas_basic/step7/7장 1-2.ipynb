## 기초

```python
sample_df= pd.DataFrame(
    [
        [9.1,1.2,1.3,1.4],
        [2.1,9.2,9.3,2.4],
        [3.1,3.2,3.3,3.4]
    ],
    index =['a','b','c'], columns=['A','B','C','D']
)
sample_df
```

```python
# 25% : 1사분위수
# 50% : 2사분위수, 중앙값
# 75% : 3사분위수
sample_df.describe()
```

```python
# 기본 옵션으로 describe() 수행시 수치형 데이터에 대한 통계값을 제공한다.
sample_df['A'].describe()
```

```python
# % 를 지정 할 수 있다.
sample_df.describe(percentiles=[.01,.25,.75,.99])
```

```python
sample_df2= pd.DataFrame(
    [
        ['a',1.2,1.3,1.4],
        ['c',9.2,9.3,2.4],
        ['d',3.2,3.3,3.4],
        ['c',4.7,2.3,5.4]
    ],
    index =['a','b','c','d'], columns=['A','B','C','D']
)
sample_df2
```

```python
# 기본 옵션으로 수행시 수치형 열에 대한 통계 자료만 보여준다.
sample_df2.describe()
```

```python
# 문자열에 대한 통계값도 구할 수 있다.
sample_df2
```

```python
# 문자열로 구성된 열에 대한 통계값을 보여준다.
sample_df2.describe(include=['object'])
# count: 전체 관측값
# unique: 유일값
# top: 최대값 (가장 많은 빈도를 가진 값) => a : 1, d : 1, c : 2
# freq: 최빈 데이터의 실제 값 (Top의 개수)
```

### 4분위수
quantile 은 수치 데이터를 크기 순서로 정렬했을 때 0% ~ 100% (0 ~ 1) 위치에 해당하는 숫자를 리턴해 준다.  
즉, Q1 은 25% , Q2 는 50%, Q3 는 75%, Q4 는 100%를 의미

```python
num = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 순서가 바뀌어도 정렬된 값으로 통계값을 보여준다.
num.describe()
```

## 평균, 중앙값, 최빈값의 관계

```python
HeroInfo = pd.read_csv("../data/HeroInfo.csv",encoding='utf-8')

for i in range(0,748) :
    a = HeroInfo.iloc[i].str.replace('-','NaN')
    HeroInfo.iloc[i] = a

HeroInfo = HeroInfo.fillna('NaN')
cols = ['Character Score', 'Intelligence', 'Strength', 'Speed','Durability', 'Power', 'Combat']
for c in cols :
    HeroInfo[c] = HeroInfo[c].astype('float')
    HeroInfo[c] = HeroInfo[c].fillna(0)
    
aa = HeroInfo.filter(items=[ 'Character Score', 'Intelligence',
                           'Strength','Speed','Durability','Power','Combat'])
```

```python
aa.head()
```

```python
# 통계값 집계
aa.describe()
```

```python
# 열이 많은 것 같아 전치 행렬로 표시
# 열이 많은 경우에 전치로 보면 .. 한 화면에 모든 열을 볼 수 있다
aa.describe().T
```

```python
import numpy as np
```

```python
# include에 [np.number] 값을 지정하면 문자열중 숫자로 열을 자동으로 식별함
HeroInfo.describe(include=[np.number]).T
```

```python
# np.object는 문자열의 열을 자동식별함. 하지만 deprecated 예정임
HeroInfo.describe(include=[np.object]).T.head()
```

```python
# 메소드 체이닝을 이용하여 평균값의 결과를 소수점 3자리에서 반올림
ch_sco = aa.groupby('Character Score').mean().round(3)
```

```python
ch_sco.shape
```

```python
ch_sco.head()
```

```python
# Power의 값을 기준으로 모든 열에 대한 평균값
HeroInfo.pivot_table(index='Power',aggfunc=np.mean).head()
```

```python
# 두개 열을 기준으로 나머지 열에 대한 평균값
HeroInfo.pivot_table(index=['Power','EyeColor'],aggfunc=np.mean).head()
```

```python
# 두개 열을 기준으로 나머지열에 대한 통계정보
HeroInfo.pivot_table(index=['Power','EyeColor'],aggfunc='describe').head()
```

```python
# 분석 정보가 많기 때문에 통계정보를 하나의 열로 선택
HeroInfo.pivot_table(index=['Power','EyeColor'],aggfunc='describe')['Character Score'].head()
```
