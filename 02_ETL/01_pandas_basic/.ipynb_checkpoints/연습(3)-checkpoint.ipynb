```python
import pandas as pd
import numpy as np
```

```python
person_data = {
    "name" : ["john","eva","ben","alice","jenny","risa"],
    "age" : [33,18,20,28,28,24],
    "job" : ["teacher","student","student","doctor","developer","actor"]
}

person = pd.DataFrame(person_data)
person
```

```python
person['salary'] = 0

person
```

```python
person['salary'] = np.where(person['job'] != 'student', 'yes', 'no')

person
```

```python
student_data = {
    "name" : ["john","eva","ben","alice","jin"],
    "midterm" : [79,98,83,84,66],
    "final" : [82,94,81,90,73]
}

student = pd.DataFrame(student_data)
student
```

```python
student['total'] = student['midterm'] + student['final']

student
```

```python
student['avg'] = student['total'] / 2

student
```

```python
grade_list = []

for grade in student['avg']:
    if grade >= 90:
        grade_list.append('A')
    elif grade >= 80 :
        grade_list.append('B')
    else:
        grade_list.append('C')
        
student['grade'] = grade_list

student
```

```python
def pass_fail(grade):
    if grade != 'C':
        return "Pass"
    else:
        return "Fail"
 
student.grade = student.grade.apply(pass_fail)

student
```
