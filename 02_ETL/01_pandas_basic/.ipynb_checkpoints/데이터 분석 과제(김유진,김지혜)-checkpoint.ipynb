# Best Music 선정

'2018 Best Music' 신규 열 추가  
조건 => Rating >= 9 and Music Count >= 10

```python
music.loc[(music['Rating'] >= 8) & (music['Music Count'] >= 9)]
```

```python
best_music = (music['Rating'] >= 8) & (music['Music Count'] >= 9)
best_music
```

```python
music['best music'] = ""
```

```python
music
```

```python
np.where(best_music, 'Best', '-')
```

```python
music['best music'] = pd.Categorical(np.where(best_music, 'Best', '-'))
music.head(55)
```

# replace 함수를 활용한 공급사 수정

공급사에서 (주)지니뮤직을 (주)멜론으로 수정

```python
music_distribution = music.copy()
music_distribution.head()
```

```python
music_distribution.loc[(music_distribution['Distribution'] == '(주)지니뮤직')]
```

```python
music_distribution.replace('(주)지니뮤직','(주)멜론')
```
