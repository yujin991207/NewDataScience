#### 새로운 데이터 생성 (2022, 2021 kbo 선수기록)

```python
player_data = {
    "선수": ["이대호", "이대호", "이정후", "이정후", "구자욱", "구자욱", "김현수", "김현수"],
    "연도": ["2022", "2021", "2022", "2021", "2022", "2021", "2022", "2021"],
    "타율": [0.304, 0.362, 0.300, 0.256, 0.288, 0.305, 0.232, 0.221],
    "홈런" : [20,25,20,23,18,5,22,13],
    "도루": [0,1,10,8,5,7,8,9],
    "소속팀": ["롯데", "롯데", "키움", "키움", "삼성", "삼성", "LG", "LG"]
}
columns = ["선수", "연도", "타율","도루","홈런","소속팀"]
player_d = pd.DataFrame(player_data, columns=columns)
player_d
```

#### 소속팀별 연도별 타율

```python
player_hit = player_d.pivot_table(values = '타율', 
                                  columns = '소속팀', 
                                  index = '연도', 
                                  aggfunc = 'sum'
                                  )
```

```python
player_hit
```

#### 소속팀별 연도별 홈런 갯수

```python
team_homerun = player_d.pivot_table(values = '홈런', 
                                    columns = '소속팀', 
                                    index = '연도', 
                                    aggfunc = 'sum')
```

```python
team_homerun
```

#### 연도별 선수별 도루 갯수

```python
stolen_base = player_d.pivot_table(values = '도루', 
                                    columns = '선수', 
                                    index = '연도', 
                                    aggfunc = 'sum')
```

#### 2021년도 선수기록 stack/unstack 사용

```python
last_year = player_d.query('연도 == "2021"')
```

```python
last_year
```

```python
last_year.pivot_table(index = '선수', columns = '연도' , values = '타율')
```

```python
last_year_i = last_year.set_index(['연도','선수'])
```

```python
last_year_i
```

```python
last_year_st = last_year_i.stack()
```

```python
last_year_st
```
