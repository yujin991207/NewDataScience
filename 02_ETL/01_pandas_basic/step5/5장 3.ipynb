## 예제 5-11  데이터를 병합 처리하기

```python
data = {

"student_id": [11111, 11112, 11113, 11114],

"Math": [60, 80, 90, 75],

"English": [85, 90, 65, 70],

"Class": [1, 2, 3, 4]

}



df = pd.DataFrame(data)

df
```

```python
# id_vars: 고정열, var_name: 병합할 열 이름, value_name: 병합할 열의 값 이름, value_vars: 병합할 Target 열
df2 = df.melt(id_vars = "student_id", var_name = "Subject", value_name = "Score", value_vars=['Math', 'English'])
df2
```

## 예제 5-12 열에 대한 추가

```python
import numpy as np
```

```python
df = pd.DataFrame({'A': range(1, 11), 'B': np.random.randn(10)})
```

```python
df
```

```python
df_1 = df.copy()
```

```python
# 기존의 열이 없으면 지정한 이름으로 갱신
# assign을 통해 새로운 열을 생성 할 수 있다.
df_1.assign(ln_A = lambda x: np.log(x.A))
```

```python
df_2 = df.copy()
```

```python
df_2['ln_A'] = np.log(df_2['A'])
```

```python
df_2
```
