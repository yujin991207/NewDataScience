```python
pd.merge(airport_code,country_code, left_on = "국  가", right_on = "나라 이름")
```

```python
all_country = pd.merge(airport_code,country_code, left_on = "국  가", right_on = "나라 이름").drop("나라 이름",axis = 1)
```

```python
all_country
```

```python
all_country = all_country.rename(columns = {'국  가':'국가','alpha-2':'국가코드'})
```

```python
all_country = all_country[['국가','숫자','alpha-3','국가코드','도시코드','도 시 명','공 항 명']]
```

```python
all_country
```

```python
all_country_gr = all_country.groupby(['국가','국가코드'])[['공 항 명']]
all_country_gr.count()
```

```python
all_country_airport = all_country_gr.count().rename(columns = {'공 항 명':'공항 갯수'})
```

```python
all_country_airport
```

```python
all_country_airport.isnull().sum()
```

```python
sample = pd.read_csv('../data/sample1.csv', encoding = 'cp949')
sample
```

```python
sample = sample.set_index('국가')
sample
```

```python
sample = sample.set_index(['국가','국가코드'])
```

```python
sample
```

```python
sample_inx = pd.merge(sample,all_country_airport, left_index = True, right_index = True, how = "left")
```

```python
sample_inx = sample_inx.fillna(0)
```

```python
sample_inx
# sample_inx['공항 갯수'].value.counts()
```

```python
sample_inx = sample_inx.astype({'공항 갯수':'int'})
sample_inx
```

```python
sample_inx.groupby('대륙명').sum()
```

```python
sample_inx = sample_inx.groupby('대륙명').agg(['sum','max','min'])
```
