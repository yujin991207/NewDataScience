## 예제 6-5  조인문장 SQL 문장과 비교

```python
import numpy as np
```

```python
df1 = pd.DataFrame({'키': ['봄', '여름', '가을', '겨울'],
                    '값': np.random.randn(4)})
```

```python
df1
```

```python
df2 = pd.DataFrame({'키': ['여름', '겨울', '겨울', '기타'],
                   '값': np.random.randn(4)})
```

```python
df2
```

SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;

```python
 pd.merge(df1, df2, on='키')
```

SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;

```python
pd.merge(df1, df2, on='키', how='left')
```

SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;

```python
pd.merge(df1, df2, on='키', how='right')
```

SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;

```python
pd.merge(df1, df2, on='키', how='outer')
```

## 예제 6-7  추가적인 조건절 처리 메소드

```python
# where 함수를 통해 간단한 검색 조건을 구현할 수 있다.
korea_2018.where(korea_2018['서울'] < 30).head(10)
# 30보다 큰 값은 NaN으로 처리한다
# 검색 결과는 True인 경우만 데이터의 값을 표시하고 나머지는 NaN으로 표시한다.
```

```python
# False인 곳에 Default 값을 지정할 수 있다.
korea_2018.where(korea_2018['서울'] < 30, 0.0).head()
# Default 값에 0을 할 당
```

```python
# False인 곳의 값만 보여주고 나머지는 NaN으로 처리한다.
# 조건을 만족하지않는곳만 데이터를 보여준다.
korea_2018.mask(korea_2018['서울'] < 30).head()
```

```python
korea_2018.mask(korea_2018['서울'] < 30, 0.0).head()
```

```python
# 원하는 행 추출
korea_2018.take([0],axis=0)
```

```python
# 원하는 열 추출
korea_2018.take([0],axis=1).head()
```

```python
korea_2018.take([0,3,5],axis=1).head()
```
