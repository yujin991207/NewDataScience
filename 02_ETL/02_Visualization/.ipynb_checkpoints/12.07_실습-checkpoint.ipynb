```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
sns.get_dataset_names()
```

#### 'flights' 데이터 이용

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_piv = flights.pivot('month','year','passengers')

flights_piv
```

```python
#cmap = plt.get_cmap('PiYG')
#plt.pcolor(flights_piv, cmap = cmap)
sns.heatmap(flights_piv, cmap = 'Greys', annot = True, fmt = "d")

plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['font.size'] = 10

plt.xticks(np.arange(0.5, len(flights_piv.columns), 1), flights_piv.columns)
plt.yticks(np.arange(0.5, len(flights_piv.index), 1), flights_piv.index)

plt.title('flights data', fontsize = 13)

plt.xlabel('Year', fontsize = 10)
plt.ylabel('Month', fontsize = 10)

#plt.matshow(flights_piv, cmap = cmap)
#plt.colorbar()
plt.show()
```

#### 'exercise' 데이터 이용

```python
exercise = sns.load_dataset('exercise')

exercise.head()
```

```python
exercise_running = exercise.loc[exercise['kind'] == 'running',:]

exercise_running.head()
```

```python
plt.rcParams['figure.figsize'] = (15,10)
plt.rcParams['font.size'] = 15

plt.subplot(2,1,1)
plt.title('exercise data')
sns.histplot(data = exercise, x = "pulse", hue = "kind", element = "poly")

plt.subplot(2,1,2)
plt.title('exercise-running data')
sns.histplot(data = exercise_running, x = "pulse", hue = "time", element = "poly")

plt.xlim(90,150)
plt.ylim(0,10)

plt.tight_layout()
plt.show()
```
