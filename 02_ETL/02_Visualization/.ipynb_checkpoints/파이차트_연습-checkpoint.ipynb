```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
```

```python
qu = [98, 20]
yes_no = ['그렇다','아니다']
explode = [0.15, 0]
colors = ['darkorange','skyblue']

plt.pie(qu, labels = yes_no, autopct = "%.1f%%", startangle = 0, 
        shadow = False, colors = colors, explode = explode)
plt.title('구로구 설문조사 결과')
plt.show()
```

```python
country = ['south Korea','USA','japan','china','spain']
ratio = [90, 77, 75,60, 33]
colors = ['darkorange','lightgray','skyblue','gold','darkgreen']
wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 3}

plt.pie(ratio, labels = country, autopct = '%.1f%%', startangle = 260, 
        colors = colors, wedgeprops = wedgeprops)
plt.show()
```

```python
data_limit = 1000000

korean = 85.50
hongkong = 87.32
japan = 87.05

variance = 4

korean_data = np.round(
    korean + 
    (np.random.randn(data_limit) * variance),1)

hongkong_data = np.round(
    hongkong + 
    (np.random.randn(data_limit) * variance),1)

japan_data = np.round(
    japan + 
    (np.random.randn(data_limit) * variance),1)
```

```python
colors = ['lightcoral','gold','black']
```

```python
plt.xlabel('수명 (세)')

plt.title('세계 평균 수명')

plt.hist((korean_data,japan_data,hongkong_data), 
         bins = 100, label = ['한국','일본','홍콩'],
         color = colors)

plt.legend()
plt.show()
```
