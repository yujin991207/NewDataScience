```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_piv = flights.pivot('month','year','passengers')

flights_piv
```

```python
# heatmap by plt.pcolor()
cmap = plt.get_cmap('PiYG')
plt.pcolor(flights_piv, cmap = cmap)

plt.xticks(np.arange(0.5, len(flights_piv.columns), 1), flights_piv.columns)
plt.yticks(np.arange(0.5, len(flights_piv.index), 1), flights_piv.index)

plt.title('Heatmap by plt.pcolor()', fontsize=15)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Month', fontsize=14)

#plt.matshow(flights_piv, cmap = cmap)
plt.colorbar()
plt.show()

```

```python
plt.style.use('default')

x = np.arange(0,4)
y1 = x ** 2
y2 = [4, 3, 2, 1]

plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['font.size'] = 12

fig, ax1 = plt.subplots()
ax1.plot(x, y1, color = 'darkorange')
ax1.set_xlabel('x')

ax2 = ax1.twinx()
ax2.plot(x, y2, color = 'darkgreen')
ax2.set_ylabel('y')

plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] =(6,3)
np.random.seed(0)

a = np.random.normal(2,3.0,100)
b = np.random.normal(-1,2.0,2000)
c = np.random.normal(3,1.5,100)

fig, ax = plt.subplots()

ax.boxplot([a, b, c])
ax.set_ylim(-10.0, 10.0)

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
