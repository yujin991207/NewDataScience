```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = sns.load_dataset("glue")
df.shape
```

```python
df.head(10)
```

```python
glue = df.pivot("Model", "Task", "Score")
```

```python
glue.shape
```

```python
glue.head(8)
```

```python
sns.heatmap(glue)
```

```python
# annot = True로 설정하게되면 각 히트맵 셀의값을 표시해준다
sns.heatmap(glue, annot=True)
```

```python
sns.heatmap(glue, annot=True, fmt=".1f")
```

```python
# 행단위로 각 셀의 값의 순위를 매김 (기본 동작은 낮은 값부터 순위를 매김)
sns.heatmap(glue, annot=glue.rank(axis="columns"))
```

```python
# 행단위로 각 셀의 값의 순위를 매김 (ascending=False로 내림차순 정렬)
sns.heatmap(glue, annot=glue.rank(axis="columns", ascending=False))
```

```python
# 열단위로 순위를 오름차순 정렬
sns.heatmap(glue, annot=glue.rank())
```

```python
sns.heatmap(glue, annot=glue.rank(ascending=False), linewidth=1)
```

```python
# 다른 컬러맵 적용
# 디폴트 컬러맵은 높은 값일 수록 진하게 또는 강하게 표현되는 색깔톤과 어울리지 않는다.
sns.heatmap(glue, annot=glue.rank(), linewidth=1, cmap='crest')
```

```python
# x축 위치 변경
# 권장사항 셀에 표현되는값을 실제값으로, 컬러맵은 수치가 높을수록 진한색, 
# 수치가 낮을수록 연한색의 스타일을 사용하여 일관성을 유지
ax=sns.heatmap(glue, annot=True, linewidth=1, cmap='crest')
ax.xaxis.tick_top()
```
