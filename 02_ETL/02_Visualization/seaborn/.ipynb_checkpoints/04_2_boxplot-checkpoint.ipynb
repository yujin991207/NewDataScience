```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = sns.load_dataset("titanic")
sns.boxplot(x=df["age"])
```

```python
# class별 age에 대한 boxplot
sns.boxplot(data=df, x="age", y="class")
```

```python
# class별 생존여부별 나이에 대한 boxplot
sns.boxplot(data=df, x="age", y="class", hue="alive")
```

```python
sns.boxplot(data=df, x="fare", y="alive")
```

```python
# order 속성의 값에 따라 분류 순서를 지정
sns.boxplot(data=df, x="fare", y="alive", order=["yes", "no"])
```
