```python
student_grade1 = [95,94,94,96,95,93,92,92,94,98]
student_grade2 = [96,97,88,82,90,92,98,92,92,91]
student_grade3 = [82,83,82,85,86,84,89,91,88,93]
Month = 10
column_month = []

for i in range(1,Month + 1):
    column_month.append(f'{i}월')
    
name_list = ['홍길동','김영희','이철수']
```

```python
math_score = pd.DataFrame([student_grade1, 
                           student_grade2, 
                           student_grade3], 
                          index = name_list, 
                          columns = column_month)
```

```python
math_score
```

```python
font1 = {'weight' : 'bold', 
         'size' : 10}
```

```python
for i in range(3):
    plt.plot(range(1,11), 
             math_score.iloc[i],
             label = math_score.index[i], 
             linestyle = 'solid', 
             marker = '*')
    
plt.axis([0,11,75,100])
plt.legend(loc = 'best', fontsize = 10)
plt.xlabel('월', loc = 'left', fontdict = font1)
plt.ylabel('점수', labelpad = 10, fontdict = font1)
plt.title('a반 월 별 수학점수', size = 13)
plt.legend(shadow = True)
plt.show()
```

```python
sex_range = ['남성','여성']
age_range =['10대','20대','30대','40대','50대']
house_range = ['1인가구','다인가구']
why_range = ['요리하기귀찮아서','스포츠 경기를 볼때',
          '친구들과 모임을 가질때','밥말고 다른게 먹고싶어서']

MAX_RECORD = 10000000

sex_list = []
age_list = []
house_list = []
why_list = []

for i in range(MAX_RECORD):
    sex_list.append(sex_range[random.choices(range(0,len(sex_range)),
                                            weights = [0.55, 0.45])[0]])
    age_list.append(age_range[random.choices(range(0,len(age_range)),
                                            weights = [0.15, 0.4, 0.2,0.125,0.125])[0]])
    house_list.append(house_range[random.choices(range(0,len(house_range)),
                                                weights = [0.75, 0.25])[0]])
    why_list.append(why_range[random.choices(range(0,len(why_range)),
                                            weights = [0.5, 0.15, 0.05, 0.3])[0]])
```

```python
delivery_food = pd.DataFrame(
    {
    "성별" : sex_list,
    "나이대" : age_list,
    "가구" : house_list,
    "이유" : why_list
    }
)

delivery_food.head()
```

```python
colors = ['darkgray']
```

```python
age = np.arange(len(delivery_food['나이대'].value_counts().index))
plt.ticklabel_format(axis = 'x',useOffset = False, style = 'plain')

plt.barh(age,delivery_food['나이대'].value_counts(), color = colors)
#plt.yticks(age,age2)
plt.yticks(age,delivery_food['나이대'].value_counts().index)
plt.title('나이대 별 배달음식 선호도 조사')
```

```python
why = np.arange(len(delivery_food['이유'].value_counts().index))
plt.ticklabel_format(axis = 'x',useOffset = False, style = 'plain')

plt.barh(why,delivery_food['이유'].value_counts(), color = colors)
plt.yticks(why,delivery_food['이유'].value_counts().index)
plt.title('배달음식을 먹는 이유')
```

```python
house = np.arange(len(delivery_food['가구'].value_counts().index))
plt.ticklabel_format(axis = 'y',useOffset = False, style = 'plain')

plt.bar(house,delivery_food['가구'].value_counts(), color = colors, width = 0.7)
plt.xticks(house,delivery_food['가구'].value_counts().index)
```
