```python
### item_price 전처리 코드 구현
```

```python
order_count = chipo.groupby('item_name')['order_id'].count()
order_count[:10]
```

### item당 주문 총량 (quantity로)

### 주문당 평균 계산금액

```python
chipo.groupby('order_id')['item_price'].sum().mean()
```

### 한 주문에 10달러 이상 사용한 id

```python
# 한 주문에 10달러 이상 사용한 id를 출력
chipo_orderid_group = chipo.groupby('order_id').sum()
chipo_orderid_group
```

```python
results = chipo_orderid_group[chipo_orderid_group.item_price >= 20]

print(results[:10])
```

```python
results2 = chipo_orderid_group[chipo_orderid_group.item_price < 20]

print(results2[:10])
```
