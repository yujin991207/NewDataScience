#### 2017~18년까지 가장 많은 히트곡을 배출한 제작사는?

```python
music_d = pd.DataFrame(musicData.groupby('Distribution')['Rating'].mean())

music_d
```

```python
music_d.reset_index(inplace = True)
```

```python
music_d = music_d.sort_values(by = 'Rating', ascending = False)

music_d.round(2).head()
```

#### 2017년 가장 많은 히트곡을 배출한 가수? [Rating 높은순]

```python
best_music_2017 = musicData.loc[(musicData['Year'] == 2017)]

best_music_2017
```

```python
best_music_2017 = best_music_2017.groupby('Artist Name')['Rating'].mean()

best_music_2017 = pd.DataFrame(best_music_2017)
best_music_2017 = best_music_2017.round(2)

best_music_2017 = best_music_2017.sort_values(by = 'Rating')

best_music_2017.tail()
```

#### 2018년 가장 많은 히트곡을 배출한 가수? [Rating 높은순]
