```python
worst_music = musicData.groupby(['Artist Name', 'Distribution','Genre'])['Rating'].mean()
#not_favorite.to_frame()
worst_music = pd.DataFrame(worst_music)
```

```python
worst_music = worst_music.sort_values(by = 'Rating')
worst_music.reset_index(inplace = True)

worst_music = worst_music.head(5)
```

```python
worst_music
```

#### 가장 인기없는 장르는 무엇인가?

```python
#musicData['Genre'].unique()
```

```python
music_g = musicData.groupby(['Genre','Distribution','Artist Name'])['Rating'].mean()

music_g = pd.DataFrame(music_g)

music_g = music_g.sort_values(by = 'Rating', ascending = False)
music_g.reset_index(inplace = True)

music_g = music_g.round(2).tail()
music_g
```

```python
result = pd.merge(worst_music, music_g, how = 'outer')

result
```
