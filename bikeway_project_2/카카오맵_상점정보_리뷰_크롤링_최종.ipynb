```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer  # 피체 벡터화
from sklearn.metrics.pairwise import cosine_similarity  # 코사인 유사도

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import glob
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import re
```

```python
input_path = "data/*202206.csv"
file_list = [f for f in glob.glob(input_path)]
matzip=pd.DataFrame()

for file in file_list:
    temp = pd.read_csv(file,encoding='utf-8',usecols =['상호명', '상권업종대분류명','상권업종중분류명',
                                '상권업종소분류명','도로명','경도', '위도'])
    matzip=matzip.append(temp)
    
matzip=matzip[matzip['상권업종대분류명']=="음식"]
matzip=matzip[matzip['상권업종중분류명']!="제과제빵떡케익"]
matzip=matzip[matzip['상권업종중분류명']!="음식배달서비스"]
matzip=matzip[matzip['상권업종중분류명']!="유흥주점"]


matzip.reset_index(inplace=True)
matzip.drop(['index'],axis=1,inplace=True)
matzip.drop(['상권업종대분류명'],axis=1,inplace=True)

matzip['kakao_keyword'] = matzip['상호명'] + " " + matzip['도로명']
matzip.to_csv("matzip_kakao.csv", index = False)
```

```python
def text_processing(text): # 한글 영어만 나오도록 설정(예:ㅋㅋㅋ 가능, ㅋㅋㅋ 뺄 거면 ㄱ-ㅣ 삭제하면 됨)
    special = re.compile(r'[^ A-Z a-z 0-9 ㄱ-ㅣ가-힣+]')
    result = special.sub('',text)
    return result
```

```python
driver = webdriver.Chrome(r'C:/Users/user/Desktop/dd/Data_Handling/chromedriver.exe')

review_info_list=[] # 리뷰를 담을 리스트 변수
stores_info_list=[] # 상점들의 정보 담을 리스트 변수

#실제 데이터 구축시 matzip['naver_keyword']로 돌려야함
#상점 하나씩 검색
for key in a['kakao_keyword']: 
    driver.get("https://map.kakao.com/") # 카카오 지도 접속하기
    searchbox = driver.find_element(By.XPATH,"//input[@id='search.keyword.query']") # 검색창에 입력하기
    searchbox.send_keys(key)#검색창에 입력하기

    searchbutton = driver.find_element(By.XPATH,"//button[@id='search.keyword.submit']") # 검색버튼
    driver.execute_script("arguments[0].click();", searchbutton)#검색버튼 누르기
    time.sleep(1)

    if len(driver.find_elements(By.XPATH,"//a[@class='moreview']")) != 0: #검색결과가 나왔을 경우 실헹
        store_info_list=[]# 상점 정보 담을 리스트변수

        driver.execute_script('window.open("about:blank", "_blank");') # 새 탭 열기
        reviewbutton = driver.find_element(By.XPATH,"//a[@class='numberofscore']")# 리뷰 버튼

        content_url = reviewbutton.get_attribute("href") #리뷰 카테고리에 저장된 리뷰 링크 가져오기
        tabs = driver.window_handles
        driver.switch_to.window(tabs[1]) # 새 탭으로 이동
        driver.get(content_url) # 링크 열기
        time.sleep(3)


        while True:
            try:
                if (driver.find_element(By.CSS_SELECTOR,'div.evaluation_review>a.link_more').text=='후기 더보기'): #후기 더보기가 나타나지 않을 때 까지 후기 더보기 클릭
                    btn_more = driver.find_element(By.CSS_SELECTOR,'div.evaluation_review>a.link_more')
                    btn_more.click()
                    time.sleep(3)
                else:
                    break
            except:
                break

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')#파싱정보 가져오기





        store_name=key #상점이름
        store_name=text_processing(store_name)
        average_rate=float(soup.select_one('a.link_evaluation>span.color_b').get_text().rsplit('점')[0])#상점점수
        review_num=soup.select_one('a.link_evaluation>span.color_g').get_text()#리뷰수
        review_num=int(review_num[1:review_num.find(')')])#리뷰수 전처리
        location='정보없음'#위치정보
        time_info='정보없음'#운영시간정보
        tel='정보없음'#번호
        review_sum=''


        for i in range(len(soup.select('div.location_detail'))):#상세정보가 있는 구역으로 이동
            if (soup.select('h4.tit_detail')[i].get_text()=='위치'):
                location=' '.join(soup.select_one('div.location_detail').get_text().replace('\n','').split())
                location=location[:location.find('(')]
            elif(soup.select('h4.tit_detail')[i].get_text()=='운영시간 안내'):
                time_info=soup.select_one('ul.list_operation').get_text().replace('\n','').rsplit('더보기')[0]
            elif(soup.select('h4.tit_detail')[i].get_text()=='연락처'):
                tel=soup.select_one('span.txt_contact').get_text().replace('\n','')

        

        review_lists = soup.select('.list_evaluation > li')#리뷰 크롤링하러 가기

        if len(review_lists) != 0 :#리뷰가 1개 이상일 때
            for review in review_lists:
                star=review.select_one('div.grade_star>span>span').get('style')
                star=int(int(star[star.find(':')+1:star.find('%')])/20) #유저 한 명이 준 별점
                review_content=review.select_one('p.txt_comment').get_text().rsplit('더보기')[0]#유저 한 명이 쓴 후기
                review_content=text_processing(review_content) 
                name=review.select_one('div.unit_info').get_text().split('\n')[2]#유저 한 명의 닉네임
                name=text_processing(name)
                date=review.select_one('div.unit_info').get_text().split('\n')[-2]#유저 한 명이 작성한 날짜
                review_sum+=' '+review_content
                member_list=[key,name, date, review_content, star]#유저 한 명이 남긴 후기를 member_list에 담아서
                review_info_list.append(member_list)#여러명의 리뷰정보가 들어있는 review_info_list에 더해준다
        
        store_info_list=[store_name,average_rate,review_num,location,time_info,tel,review_sum]#store_info_list에 상점 하나의 정보 담아서
        stores_info_list.append(store_info_list)#여러개의 상점 정보가 들어있는 stores_info_list에 더해준다
        
        driver.close()        
        driver.switch_to.window(tabs[0])#검색창으로 이동
        
    else:
        continue
```
