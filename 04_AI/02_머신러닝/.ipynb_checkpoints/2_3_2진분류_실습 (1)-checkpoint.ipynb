```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
heart = pd.read_csv('heart.csv', sep=',', header=0)
```

```python
heart.columns
```

```python
heart_data = [[q,w,e,r,t,y,u,i,o,p,a,s,d] for q,w,e,r,t,y,u,i,o,p,a,s,d in zip(heart['age'], heart['sex'],heart['cp'],heart['trestbps'],heart['chol'],heart['fbs'],heart['restecg'],heart['thalach'],heart['exang'],heart['oldpeak'],heart['slope'],heart['ca'],heart['thal'])]
```

feature

```python
heart_data
```

label

```python
heart_label=np.column_stack((heart['target'], ))
```

## 사이킷런

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
len(heart)
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    heart_data, heart_label,stratify=heart_label, random_state=300)
```

```python
print(train_target.shape, test_target.shape)
```

```python
print(len(train_input), len(test_input))
```

```python
print(test_target)
```

# 표준화

```python
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
print(mean, std)
```

```python
train_scaled=(train_input-mean)/std
```

```python
test_scaled = ((test_input) - mean) / std
```

# 학습

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(train_input, train_target)
```

```python
kn.fit(train_scaled, train_target)
```

# test

```python
kn.score(test_scaled, test_target)
```
