## AJax 통신 정보 확인하기

### Step 1] 원하는 데이터 여부 확인
* Network - Fetch/XHR - Name
* Request 별 오른쪽탭에 Preview를 보면서 원하는 데이터를 찾는다.

### Step 2] 실제 실시간 통신 URL 정보 확인
* Header의 Request URL 정보 확인

### Step 3] user-agent 정보 확인
* Console에서 navigator.userAgent 검색

```python
custom_header ={
    'referer' : 'https://finance.daum.net/domestic',
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}

url = 'https://finance.daum.net/api/search/ranks?limit=10'
requests.get
req = requests.get(url, headers = custom_header)
# req = requests.get(url)
```

```python
print(req.status_code)
if req.status_code == requests.codes.ok:
    print("접속 성공")
    realtime_data = json.loads(req.text)
    print(realtime_data)
```
