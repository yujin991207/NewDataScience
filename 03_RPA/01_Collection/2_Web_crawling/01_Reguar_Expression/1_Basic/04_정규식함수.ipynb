# 정규식 함수
## 파이썬에서 제공하는 정규식 함수

# match

```python
import re
```

```python
p = re.compile('[a-z]+')
m = p.match('python')
print(m)

if m:
    print('Match found:',m.group()) # group() 매치된 문자열 반환
else:
    print('No match')

p = re.compile('[a-z]+')
m = p.match('3 python')
print(m)
```

```python
p = re.compile('[a-z]+')
m = p.match('3 python') # 정규식을 첫번째 글자부터 적용
print(m)
m = p.search('3 python')
print(m)
```

# findall

```python
original_text='life is too short'
p=re.compile('[a-z]+')

m = p.search(original_text) # 첫번째 매치되는 결과를 반환
print(m)

match_list = p.findall(original_text)
# 검색 결과는 매칭된 문자열들을 리스트로 반환한다.
print(match_list)
for match_element in match_list:
    print(match_element)
```

# finditer
### match 오브젝트에서 제공하는 함수

```python
original_text='life is too short'
p=re.compile('[a-z]+')

result = p.finditer(original_text)
# 매칭된 결과를 'Match object' 리스트로 반환한다. 매치 오브젝트에 대한 부가 정보를 얻을 수 있다.
for r in result:
    print(r)
    print(f'r.group(): {r.group()}, r.start(): {r.start()}, r.end(): {r.end()}, r.span(): {r.span()}\n')
```
